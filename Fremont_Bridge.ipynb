We are using Fremont Bridge dataset to get familiar with libraries such as pyplot and pandas.

**The aim is to use pyplot to visualize datasets**



```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn
from google.colab import files
#using files library we can upload local files to use in the notebook.
uploaded = files.upload()

data = pd.read_csv("fremont-bridge.csv",parse_dates=True)
 
#adding a coloumn to get total cycles coming from east and west
data['Total'] = data['West']+data['East']
#print(data.head())

#To view different details of the data
#print(data.dropna().describe())

#Here we are converting Date coloumn to datetime format so that we can extract h,m,s
data['Date']= pd.to_datetime(data['Date'])
print(data.head())

data['Hour'] = data['Date'].dt.hour
data['Day'] = data['Date'].dt.day
data['Month'] = data['Date'].dt.month
data['Year'] = data['Date'].dt.year
data['Dayname'] = data['Date'].dt.day_name()


print(data.head())


```

Now we will use pyplot and seaborn to visualize the data 

```

# Hourly movement of cyclist 
'''
In this plot we can observe that the peaks correspond to the 
school/office timings. 

'''

hourly = data.groupby('Hour').mean()['East']
hourly.plot(style=[':'],legend=True)
hourly = data.groupby('Hour').mean()['West']
hourly.plot(style=['--'],legend=True)

```

```
# Daily movement of cyclists  
'''

'''

daily = data.groupby('Day').mean()['East']
daily.plot(style=[':'],legend=True)
daily = data.groupby('Day').mean()['West']
daily.plot(style=['-'],legend=True)
```

```
# Monthly movement of cyclists
'''
This plot is in sync with the seasons throughout the year.
Summer bringing out the most number of cyclists around mid year.
'''
monthly = data.groupby('Month').mean()['East']
monthly.plot(style=[':'],legend=True)
monthly = data.groupby('Month').mean()['West']
monthly.plot(style=['-'],legend=True)
```

```
#  Movement of cyclists on different days 
'''
Here as expected people rest on Sat and Sun 
and hence there less cyclists out.
'''
dayname = data.groupby('Dayname',sort=False).mean()['East']
print(dayname)
dayname.plot(style=[':'],legend=True)
dayname = data.groupby('Dayname',sort=False).mean()['West']
dayname.plot(style=['-'],legend=True)
```
